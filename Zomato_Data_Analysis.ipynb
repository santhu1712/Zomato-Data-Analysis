```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("/content/Zomato data .csv")
print(dataframe.head())
```

```python
def handleRate(value):
	value=str(value).split('/')
	value=value[0];
	return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'], palette="viridis")
plt.xlabel("Type of Restaurant")
plt.ylabel("Count")
plt.title("Distribution of Restaurant Types")
plt.show()
```

```python
# Grouping the data and creating the DataFrame
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'Type of Restaurant': grouped_data.index, 'Votes': grouped_data.values})

# Plotting the data
plt.figure(figsize=(6, 4))  # Adjust figure size for better visualization
plt.plot(result['Type of Restaurant'], result['Votes'], color="green", marker="o", linestyle="--", linewidth=2, markersize=8)

# Adding labels and title
plt.xlabel("Type of Restaurant", color="red", fontsize=9)
plt.ylabel("Votes", color="red", fontsize=9)
plt.title("Total Votes by Type of Restaurant", fontsize=10, color="blue")
plt.xticks(rotation=45, fontsize=12)  # Rotate x-axis labels for better readability
plt.grid(alpha=0.3)  # Add grid lines for visual aid

# Display the plot
plt.tight_layout()
plt.show()
```

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'name']

print("Restaurant with the maximum votes:")
print(restaurant_with_max_votes)
```
